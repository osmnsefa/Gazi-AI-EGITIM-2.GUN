```python
import cv2
import time
import mediapipe as mp
```

```python
# EER VDEO ZERNDEN DENEMEK STERSEK VDEOYU E AKTARIP CAP'E ETLYORUZ
cap=cv2.VideoCapture(0) # 0 demek mevcut kamera

mpHand=mp.solutions.hands

hands=mpHand.Hands(max_num_hands=1) #bunun parametrelerini optimize etmek algoritmamzn hzn arttrr.

mpDraw = mp.solutions.drawing_utils



while True:
    success,img=cap.read()
    imgRGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
    results = hands.process(imgRGB)
    print(results.multi_hand_landmarks) # Elin ekrandaki konumu (eklemlerin)
    
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            mpDraw.draw_landmarks(img,handLms,mpHand.HAND_CONNECTIONS)
            
            for id, lm in enumerate(handLms.landmark):
                
                h,w,c=img.shape
                
                cx,cy =int(lm.x*w),int(lm.y*h) # Koordinatlar
                
                # Elimzden bir eklemin id'sini seiyoruz.(hepsinin bir id'si var)
                # # isaret uc = 8
                if id == 8:
                     cv2.circle(img, (cx,cy), 9, (255,0,0), cv2.FILLED)
                # # isaret uc = 6
                if id == 6:
                     cv2.circle(img, (cx,cy), 9, (0,0,255), cv2.FILLED) 
   
    
    cv2.imshow("img",img)
    if cv2.waitKey(1) & 0xFF == ord("q"): break
cap.release() #serbest brak
cv2.destroyAllWindows()
```
